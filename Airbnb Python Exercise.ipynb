# NYC Airbnb Data Analysis

Dataset: https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data

### Context
Since 2008, guests and hosts have used Airbnb to expand on traveling possibilities and present more unique, personalized way of experiencing the world. This dataset describes the listing activity and metrics in NYC, NY for 2019.

### Content
This data file includes all needed information to find out more about hosts, geographical availability, necessary metrics to make predictions and draw conclusions.

### Acknowledgements
This public dataset is part of Airbnb, and the original source can be found on this website: http://insideairbnb.com/.

```python
# Importing the necessary libraries to be used later in this exercise
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Part 1: Understanding the dataset

```python
# Downloading the raw dataset
data_raw = pd.read_csv('https://raw.githubusercontent.com/kriwang27/PythonProjs/main/AB_NYC_2019.csv')
# Reviewing the first few rows of the data
data_raw.head()
```

```python
# Getting a better view of the column headings
data_raw.columns
```

```python
# Understanding the datatypes of each of the columns, some might need to be changed for deeper analysis
data_raw.dtypes
```

## Part 2: Cleaning the data

```python
# Drop columns we won't use
# (keeping this analysis simple, I won't be looking at exact locations by latitude or longitude.)
# (I also will not be using the last review as there is no information about when this last review was recorded
# and will not add to our analysis at this time.)
data_dropped_columns = data_raw.drop(columns=['latitude', 'longitude', 'last_review'], axis=1)
data_dropped_columns
```

```python
# Drop rows with NA/no data in ALL columns

data_clean = data_dropped_columns.dropna(axis=0)
data_clean
```

```python
# Understanding the basic statistics for all columns
# Note: This won't make sense for any categorical/non-numeric columns of data, such as the name, room_type, etc.
data_clean.describe()
```

## Part 3: Begin to Analyze the data

### Average Price per Borough

```python
# Information by NYC Borough
data_clean.groupby('neighbourhood_group', sort=False).mean()
```

```python
# Visualize the average price per borough
sns.barplot(x='neighbourhood_group', y='price', data=data_clean).set_title('Average Price per Borough')
# It is not surprising that Manhattan and Brooklyn are the two highest average prices
```

### Average Price per Room Type

```python
# Breakdown of data by room type
data_clean.groupby('room_type', sort=False).mean()
```

```python
# Visualization of average price by room type
sns.barplot(x='room_type', y='price', data=data_clean).set_title('Average Price per Room Type')
# Given what I know about Airbnb's, this price breakdown by room type is not surprising.
```

### Average Price per Room Type within each Borough

```python
# Visualization
sns.barplot(x='neighbourhood_group', y='price', hue='room_type', data=data_clean).set_title('Average Price per Room Type within each Borough')
```

## Part 4: Brooklyn and Manhattan Analysis

As an alumni of NYU, most of my experiences were in Brooklyn and Manhattan. Given my knowledge and familiarity of the two neighborhoods, I want to focus the rest of my analysis in these areas. 

In the next few analyses, I will be breaking down the data by the two boroughs and doing a deeper dive into each one. 

### Average Price of Each Neighborhood within only Brooklyn and Manhattan

```python
# Avg price for each neighbourhood within Brooklyn and Manhattan

# Creating a new dataset for only Brooklyn and Manhattan rows of data
bk_man = ['Brooklyn','Manhattan']
data_clean.neighbourhood_group.isin(bk_man)
```

```python
data_bk_man = data_clean[data_clean.neighbourhood_group.isin(bk_man)]
data_bk_man
```

```python
# Sorting the Brooklyn-and-Manhattan-Only dataset by their subsequent Boroughs and then Neighbourhoods
data_bk_man_grouped = data_bk_man.groupby(['neighbourhood_group', 'neighbourhood'], sort=True).mean()
data_bk_man_grouped
```

```python
# Visualizing the dataset with a scatterplot of the average price per neighborhood
g=sns.catplot(x='neighbourhood', y='price', data=data_bk_man, height=10, aspect=20/10)
g.set_xticklabels(rotation=40)
# Given the number of neighborhoods, this isn't the clearest or most useful view.
# It also looks like there are several outliars in certain neighborhoods.
# I would like to remove those outliars, but first I need to view this data better to understand where the outliers are coming from.
```

```python
# Given what I know about Brooklyn and Manhattan and the over-packed X axis in the graph above,
# I'd like to separate the two boroughs into their own datasets. 

# Brooklyn-Only Dataset
bk = ['Brooklyn']
data_clean.neighbourhood_group.isin(bk)
data_bk = data_clean[data_clean.neighbourhood_group.isin(bk)]
data_bk
```

```python
# Manhattan-Only Dataset
man = ['Manhattan']
data_clean.neighbourhood_group.isin(man)
data_man = data_clean[data_clean.neighbourhood_group.isin(man)]
data_man
```

```python
# Visualization of the Brooklyn-Only average prices per neighborhood
g=sns.catplot(x='neighbourhood', y='price', data=data_bk, height=10, aspect=20/10)
g.set_xticklabels(rotation=90) 
# Again, I see several outliars. 
```

```python
# Visualization of the Manhattan-Only average prices per neighbood
g=sns.catplot(x='neighbourhood', y='price', data=data_man, height=10, aspect=20/10)
g.set_xticklabels(rotation=90)
# Again, I see several outliars.
```

```python
# Taking out outliers to get a better view of the data

groups = data_bk_man.groupby('neighbourhood')['price']
groups_mean = groups.transform('mean')
groups_std = groups.transform('std')
m = data_bk_man['price'].between(groups_mean.sub(groups_std.mul(3)),
                      groups_mean.add(groups_std.mul(3)),
                      inclusive=False)
print(m)

# Brooklyn-and-Manhattan-Only dataset without outliars (I couldn't think of a better dataset name, sorry)
no_outies = data_bk_man.loc[m]

no_outies
```

```python
# Brooklyn-Only dataset without outliars
bk = ['Brooklyn']
no_outies.neighbourhood_group.isin(bk)
no_outies_bk = no_outies[no_outies.neighbourhood_group.isin(bk)]
no_outies_bk
```

```python
# Scatterplot of Brooklyn-Only data without outliers
g=sns.catplot(x='neighbourhood', y='price', data=no_outies_bk, height=10, aspect=20/10)
g.set_xticklabels(rotation=90) 
# this looks much better without the outliars, but still not the best for analysis
```

```python
# Here's a box and whisker plot of the Brooklyn-Only data
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.boxplot(x='neighbourhood', y='price', data=no_outies_bk) 
sns.despine()

plt.xticks(rotation=90)

# Observations:
# I'm not sure what Sea-Gate is, but it has a surprisngly wide range of prices compared to other neighbhorhoods
# Given what I know about Williamsburg and DUMBO, I'm surprised the average isn't significantly higher than others.
```

```python
# Manhattan-Only dataset without outliars
man = ['Manhattan']
no_outies.neighbourhood_group.isin(man)
no_outies_man = no_outies[no_outies.neighbourhood_group.isin(man)]
no_outies_man
```

```python
# Scatterplot of Manhattan-Only data without outliars
g=sns.catplot(x='neighbourhood', y='price', data=no_outies_man, height=10, aspect=20/10)
g.set_xticklabels(rotation=90) 
# Again, this looks much better than before, but I'd like to take a look at the box-and-whisker plot
```

```python
# Box-and-Whisker plot
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.boxplot(x='neighbourhood', y='price', data=no_outies_man) 
sns.despine()

plt.xticks(rotation=90)

# Observations
## Tribeca has a very wide range as well as a higher average than other neighborhoods.
## I'm not surprised given the higher cost of real estate in that area and more luxury hotel costs.
## (Comparing these average prices to hotel and real estate values in restpective areas would be a cool future analysis)
```
